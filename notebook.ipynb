```python
# =====================================================
# OP'26 ANALYTICS PROJECT - COMPLETE PIPELINE
# =====================================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# =====================================================
# LOAD FILES
# =====================================================

occupancy = pd.read_csv("occupancy.csv")
volume = pd.read_csv("volume.csv")
duration = pd.read_csv("duration.csv")
price = pd.read_csv("price.csv")

information = pd.read_csv("information.csv")
time = pd.read_csv("time.csv")

# =====================================================
# CONVERT WIDE -> LONG
# =====================================================

occ_long = occupancy.melt(
    id_vars=["timestamp"],
    var_name="zone",
    value_name="occupancy"
)

vol_long = volume.melt(
    id_vars=["timestamp"],
    var_name="zone",
    value_name="volume"
)

dur_long = duration.melt(
    id_vars=["timestamp"],
    var_name="zone",
    value_name="duration"
)

price_long = price.melt(
    id_vars=["timestamp"],
    var_name="zone",
    value_name="price"
)

# =====================================================
# MERGE TEMPORAL DATASETS
# =====================================================

data = occ_long.merge(
    vol_long,
    on=["timestamp", "zone"]
)

data = data.merge(
    dur_long,
    on=["timestamp", "zone"]
)

data = data.merge(
    price_long,
    on=["timestamp", "zone"]
)

# =====================================================
# ADD INFORMATION FEATURES
# =====================================================

information = information.rename(
    columns={"num": "zone"}
)

information["zone"] = information["zone"].astype(str)
data["zone"] = data["zone"].astype(str)

data = data.merge(
    information,
    on="zone",
    how="left"
)

# =====================================================
# ADD TIME FEATURES
# =====================================================

time["timestamp"] = occupancy["timestamp"]

data = data.merge(
    time,
    on="timestamp",
    how="left"
)

# =====================================================
# SORT DATA
# =====================================================

data = data.sort_values(
    ["zone", "timestamp"]
)

# =====================================================
# CREATE LAG FEATURES
# =====================================================

data["occ_lag1"] = (
    data.groupby("zone")["occupancy"]
    .shift(1)
)

data["vol_lag1"] = (
    data.groupby("zone")["volume"]
    .shift(1)
)

data["price_lag1"] = (
    data.groupby("zone")["price"]
    .shift(1)
)

data["occ_lag24"] = (
    data.groupby("zone")["occupancy"]
    .shift(24)
)

data["vol_lag24"] = (
    data.groupby("zone")["volume"]
    .shift(24)
)

data["price_lag24"] = (
    data.groupby("zone")["price"]
    .shift(24)
)

# =====================================================
# ROLLING FEATURES
# =====================================================

data["occ_roll24"] = (
    data.groupby("zone")["occupancy"]
    .rolling(24)
    .mean()
    .reset_index(level=0, drop=True)
)

data["vol_roll24"] = (
    data.groupby("zone")["volume"]
    .rolling(24)
    .mean()
    .reset_index(level=0, drop=True)
)

# =====================================================
# REMOVE NaNs
# =====================================================

data = data.dropna()

print("Final Shape:", data.shape)

# =====================================================
# TARGET
# =====================================================

TARGET = "occupancy"

y = data[TARGET]

X = data.drop(
    columns=[
        TARGET,
        "timestamp"
    ]
)

# =====================================================
# ENCODE CATEGORICAL
# =====================================================

X = pd.get_dummies(
    X,
    columns=["zone"],
    drop_first=True
)

# =====================================================
# TIME BASED SPLIT
# =====================================================

split = int(len(X) * 0.8)

X_train = X.iloc[:split]
X_test = X.iloc[split:]

y_train = y.iloc[:split]
y_test = y.iloc[split:]

# =====================================================
# MODEL
# =====================================================

try:

    from xgboost import XGBRegressor

    model = XGBRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=8,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )

except:

    from sklearn.ensemble import RandomForestRegressor

    model = RandomForestRegressor(
        n_estimators=300,
        random_state=42,
        n_jobs=-1
    )

# =====================================================
# TRAIN
# =====================================================

model.fit(
    X_train,
    y_train
)

# =====================================================
# PREDICT
# =====================================================

pred = model.predict(X_test)

# =====================================================
# EVALUATE
# =====================================================

mae = mean_absolute_error(
    y_test,
    pred
)

rmse = mean_squared_error(
    y_test,
    pred
) ** 0.5

r2 = r2_score(
    y_test,
    pred
)

print("\nRESULTS")
print("MAE :", mae)
print("RMSE:", rmse)
print("R2  :", r2)

# =====================================================
# FEATURE IMPORTANCE
# =====================================================

try:

    importance = pd.DataFrame({
        "Feature": X.columns,
        "Importance": model.feature_importances_
    })

    importance = importance.sort_values(
        "Importance",
        ascending=False
    )

    print("\nTOP 20 FEATURES")
    print(
        importance.head(20)
    )

except:
    pass
```

```python
def adjusted_volume(row):

    base_volume = row["volume"]

    tariff = row["dynamic_tariff_v2"]

    change = (tariff - 15) / 15

    elasticity = -0.30

    return base_volume * (1 + elasticity * change)
```

```python
q30 = data["occupancy_density"].quantile(0.30)
q70 = data["occupancy_density"].quantile(0.70)

print(q30, q70)
```

```python
def tariff_agent(util):

    if util <= q30:
        return 13

    elif util >= q70:
        return 22

    else:
        return 15
```

```python
data["agent_tariff"] = (
    data["occupancy_density"]
    .apply(tariff_agent)
)
```

```python
def demand_response(row):

    base = row["volume"]

    change = (
        row["agent_tariff"] - 15
    ) / 15

    elasticity = -0.6

    return base * (
        1 + elasticity * change
    )
```

```python
data["agent_volume"] = (
    data.apply(
        demand_response,
        axis=1
    )
)
```

```python
data["agent_revenue"] = (
    data["agent_volume"]
    *
    data["agent_tariff"]
)
```

```python
agent_gain = (
    (
        data["agent_revenue"].sum()
        -
        data["baseline_revenue"].sum()
    )
    /
    data["baseline_revenue"].sum()
) * 100

print(agent_gain)
```

```python
data["agent_occupancy"] = (
    data["occupancy"]
    *
    (
        data["agent_volume"]
        /
        data["volume"]
    )
)

old_occ = data["occupancy"].mean()

new_occ = data["agent_occupancy"].mean()

agent_wait_reduction = (
    (old_occ - new_occ)
    /
    old_occ
) * 100

print(agent_wait_reduction)
```

```python
import pandas as pd

occupancy = pd.read_csv("occupancy.csv")
volume = pd.read_csv("volume.csv")
duration = pd.read_csv("duration.csv")
price = pd.read_csv("price.csv")

occ_long = occupancy.melt(
    id_vars="timestamp",
    var_name="zone",
    value_name="occupancy"
)

vol_long = volume.melt(
    id_vars="timestamp",
    var_name="zone",
    value_name="volume"
)

dur_long = duration.melt(
    id_vars="timestamp",
    var_name="zone",
    value_name="duration"
)

price_long = price.melt(
    id_vars="timestamp",
    var_name="zone",
    value_name="price"
)

print(occ_long.shape)
print(vol_long.shape)
```

```python
master = (
    occ_long
    .merge(
        vol_long,
        on=["timestamp","zone"]
    )
    .merge(
        dur_long,
        on=["timestamp","zone"]
    )
    .merge(
        price_long,
        on=["timestamp","zone"]
    )
)

print(master.shape)

master.head()
```

```python
stations = pd.read_csv("stations.csv")

stations = stations.rename(
    columns={
        "station_id":"zone"
    }
)

stations["zone"] = (
    stations["zone"]
    .astype(str)
)

master["zone"] = (
    master["zone"]
    .astype(str)
)

print(stations.head())
```

```python
master = master.merge(
    stations,
    on="zone",
    how="left"
)

print(master.shape)

print(
    master[
        [
            "latitude",
            "longitude",
            "fast",
            "slow",
            "count"
        ]
    ].isnull().sum()
)
```

```python
import numpy as np

data = master.copy()

# Utilization Rate
data["utilization_rate"] = (
    data["duration"] / 24
)

# Occupancy Density
data["occupancy_density"] = (
    data["occupancy"] /
    data["count"]
)

# Revenue
data["revenue"] = (
    data["volume"] *
    data["price"]
)

# Congestion Label
data["is_congested"] = (
    data["occupancy_density"] > 0.80
).astype(int)

print(data.shape)

data[
    [
        "utilization_rate",
        "occupancy_density",
        "revenue",
        "is_congested"
    ]
].head()
```

```python
data["hour"] = (
    (data["timestamp"] - 1)
    % 24
)

data["day"] = (
    (data["timestamp"] - 1)
    // 24
)

data["weekday"] = (
    data["day"] % 7
)

data["is_weekend"] = (
    data["weekday"] >= 5
).astype(int)

print(
    data[
        [
            "timestamp",
            "hour",
            "day",
            "weekday",
            "is_weekend"
        ]
    ].head()
)
```

```python
data = data.sort_values(
    ["zone","timestamp"]
)

data = data.reset_index(
    drop=True
)

print(data.shape)
```

```python
data["occ_lag1"] = (
    data.groupby("zone")["occupancy"]
    .shift(1)
)

data["vol_lag1"] = (
    data.groupby("zone")["volume"]
    .shift(1)
)

data["dur_lag1"] = (
    data.groupby("zone")["duration"]
    .shift(1)
)

data["occ_lag24"] = (
    data.groupby("zone")["occupancy"]
    .shift(24)
)

data["vol_lag24"] = (
    data.groupby("zone")["volume"]
    .shift(24)
)

data["dur_lag24"] = (
    data.groupby("zone")["duration"]
    .shift(24)
)
```

```python
data["occ_roll24"] = (
    data.groupby("zone")["occupancy"]
    .transform(
        lambda x:
        x.rolling(
            24,
            min_periods=1
        ).mean()
    )
)

data["vol_roll24"] = (
    data.groupby("zone")["volume"]
    .transform(
        lambda x:
        x.rolling(
            24,
            min_periods=1
        ).mean()
    )
)

data["dur_roll24"] = (
    data.groupby("zone")["duration"]
    .transform(
        lambda x:
        x.rolling(
            24,
            min_periods=1
        ).mean()
    )
)
```

```python
data = data.dropna()

print(data.shape)
```

```python
# ==========================================
# UTILIZATION PREDICTION MODEL (CORRECTED)
# ==========================================

import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# ------------------------------------------
# FEATURES
# ------------------------------------------

features_util = [

    "occupancy",
    "volume",
    "price",

    "fast",
    "slow",
    "count",

    "latitude",
    "longitude",

    "hour",
    "weekday",
    "is_weekend",

    "occ_lag1",
    "vol_lag1",

    "occ_lag24",
    "vol_lag24",

    "occ_roll24",
    "vol_roll24"

]

# ------------------------------------------
# X AND Y
# ------------------------------------------

X_util = data[features_util]

y_util = data["utilization_rate"]

print("X Shape:", X_util.shape)
print("Y Shape:", y_util.shape)

# ------------------------------------------
# TRAIN TEST SPLIT
# ------------------------------------------

split = int(len(data) * 0.8)

X_train = X_util.iloc[:split]
X_test = X_util.iloc[split:]

y_train = y_util.iloc[:split]
y_test = y_util.iloc[split:]

print("Train:", X_train.shape)
print("Test :", X_test.shape)

# ------------------------------------------
# MODEL
# ------------------------------------------

util_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=15,
    random_state=42,
    n_jobs=-1
)

# ------------------------------------------
# TRAIN
# ------------------------------------------

util_model.fit(
    X_train,
    y_train
)

# ------------------------------------------
# PREDICT
# ------------------------------------------

util_pred = util_model.predict(
    X_test
)

# ------------------------------------------
# METRICS
# ------------------------------------------

util_mae = mean_absolute_error(
    y_test,
    util_pred
)

util_rmse = np.sqrt(
    mean_squared_error(
        y_test,
        util_pred
    )
)

util_r2 = r2_score(
    y_test,
    util_pred
)

print("\n==============================")
print("UTILIZATION MODEL RESULTS")
print("==============================")

print("MAE :", util_mae)
print("RMSE:", util_rmse)
print("R2  :", util_r2)

# ------------------------------------------
# FEATURE IMPORTANCE
# ------------------------------------------

importance = pd.DataFrame({

    "Feature": features_util,
    "Importance": util_model.feature_importances_

})

importance = importance.sort_values(
    "Importance",
    ascending=False
)

print("\nTOP FEATURES")
print(importance.head(20))

# ------------------------------------------
# SAVE
# ------------------------------------------

importance.to_csv(
    "utilization_feature_importance.csv",
    index=False
)

results_util = pd.DataFrame({

    "Model": ["RF_Utilization"],

    "Target": ["utilization_rate"],

    "MAE": [util_mae],

    "RMSE": [util_rmse],

    "R2": [util_r2]

})

results_util.to_csv(
    "utilization_results.csv",
    index=False
)

print("\nSaved:")
print("utilization_results.csv")
print("utilization_feature_importance.csv")
```

```python
# ==========================================
# CONGESTION CLASSIFICATION AGENT
# ==========================================

import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix
)

features_cong = [

    "occupancy",
    "volume",
    "duration",
    "price",

    "fast",
    "slow",
    "count",

    "latitude",
    "longitude",

    "hour",
    "weekday",
    "is_weekend",

    "occ_lag1",
    "vol_lag1",
    "dur_lag1",

    "occ_lag24",
    "vol_lag24",
    "dur_lag24",

    "occ_roll24",
    "vol_roll24",
    "dur_roll24"

]

X_cong = data[features_cong]

y_cong = data["is_congested"]

split = int(len(data) * 0.8)

X_train = X_cong.iloc[:split]
X_test = X_cong.iloc[split:]

y_train = y_cong.iloc[:split]
y_test = y_cong.iloc[split:]

model_cong = RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    random_state=42,
    n_jobs=-1
)

model_cong.fit(
    X_train,
    y_train
)

pred_class = model_cong.predict(X_test)

pred_prob = model_cong.predict_proba(X_test)[:,1]

acc = accuracy_score(
    y_test,
    pred_class
)

prec = precision_score(
    y_test,
    pred_class
)

rec = recall_score(
    y_test,
    pred_class
)

f1 = f1_score(
    y_test,
    pred_class
)

auc = roc_auc_score(
    y_test,
    pred_prob
)

print("\n==============================")
print("CONGESTION MODEL RESULTS")
print("==============================")

print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)
print("F1 Score :", f1)
print("AUC      :", auc)

print("\nConfusion Matrix")
print(
    confusion_matrix(
        y_test,
        pred_class
    )
)

importance_cong = pd.DataFrame({
    "Feature": features_cong,
    "Importance": model_cong.feature_importances_
})

importance_cong = importance_cong.sort_values(
    "Importance",
    ascending=False
)

print("\nTOP FEATURES")
print(
    importance_cong.head(15)
)

importance_cong.to_csv(
    "congestion_feature_importance.csv",
    index=False
)
```
